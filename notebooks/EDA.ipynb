```python
ax = plt.axes()
sns.heatmap(books.isna().transpose(), cbar=False, ax=ax, cmap='viridis')
ax.set_title('Missing values in the dataset')

plt.xlabel("Columns")
plt.ylabel("Missing values")

plt.show()
```

```python
books["missing_description"] = np.where(books["description"].isna(), 1, 0)
books["age_of_book"] = 2024 - books["published_year"]
```

```python
columns_of_interest = ["num_pages", "age_of_book", "missing_description", "average_rating"]

# 2. Handle missing data explicitly
subset_data = books[columns_of_interest].dropna()  # or use appropriate imputation

# 3. Calculate correlation matrix
correlation_matrix = subset_data.corr(method="spearman")

# 4. Create visualization
sns.set_theme(style="white")
plt.figure(figsize=(8, 6))

# Create heatmap with enhancements
heatmap = sns.heatmap(
    correlation_matrix,
    annot=True,
    fmt=".2f", 
    cmap="coolwarm",
    cbar_kws={"label": "Spearman Correlation Coefficient"},
    vmin=-1,
    vmax=1,
    mask=np.triu(np.ones_like(correlation_matrix, dtype=bool))  # Mask upper triangle
)

# Improve labels and readability
heatmap.set_title("Spearman Correlation Between Book Features", pad=20, fontsize=14, fontweight="bold")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=45, ha='right')
heatmap.set_yticklabels(heatmap.get_yticklabels(), rotation=0)
plt.tight_layout()

plt.show()
```

```python
book_missing = books[~(books["description"].isna()) &
      ~(books["num_pages"].isna()) &
      ~(books["average_rating"].isna()) &
      ~(books["published_year"].isna())
]
```

```python
book_missing_25_words = book_missing[book_missing["words_in_description"] >= 25]
```

```python
book_missing_25_words["title_and_subtitle"] = (
    np.where(book_missing_25_words["subtitle"].isna(), book_missing_25_words["title"],
             book_missing_25_words[["title", "subtitle"]].astype(str).agg(": ".join, axis=1))
)
```

```python
book_missing_25_words["tagged_description"] = book_missing_25_words[["isbn13", "description"]].astype(str).agg(" ".join, axis=1)
```
